```python
# Question 1

def timesOccursIn(k, A, i=0, total=0):
    if (i==len(A)):
        return total
    elif A[i] == k:
        return timesOccursIn(k,A,i+1,total+1)
    else:
        return timesOccursIn(k, A, i+1, total)
    
# testing

A = [56,5,31,45,31]
print(timesOccursIn(5, A))
print(timesOccursIn(31, A))
```

```python
# Question 2

def multAll(k,A):
    if len(A)==0:
        return A
    else:
        return multAllRec(k,A,0)

def multAllRec(k,A,i):
    if i==len(A):
        return A
    else:
        A[i] = A[i]*k
        return multAllRec(k,A,i+1)
    
# Testing

A = [7]
print(A)
print(multAll(10,A))
print(multAll(7,A))
    
```

```python
# Question 4

def binSearch2(A,k,lo=0,hi=len(A)-1):
    if lo <= hi:
        mid = (lo+hi)//2
        if A[mid] == k:
            return mid
        elif A[mid] < k:
            lo = mid+1
            return binSearch2(A,k,lo,hi)
        else:
            hi = mid-1
            return binSearch2(A,k,lo,hi)
    return -1

# testing

A = [3,6,9,12,45]
print(binSearch2(A,12))
print(binSearch2(A,45))
print(binSearch2(A,6))
```

```python
# Question 5

import random
import time

def randomIntArray(s,n):
    out = []
    for i in range(s):
        out.append(random.randint(0,n))
    return out

def sortTimeUsing(sortf,A):
    t = time.time()
    sortf(A)
    t = time.time()-t
    return t

# iterative
def selectionSort(A):
    for i in range(len(A)):
        imin = findMin(i,A)
        swap(i,imin,A)
        
def findMin(i, A):
    imin = i
    for j in range(i+1,len(A)):
        if A[j] < A[imin]: 
            imin = j
    return imin

def swap(i, j, A):
    tmp = A[i]
    A[i] = A[j]
    A[j] = tmp
    
def insertionSort(A):
    for i in range(1,len(A)):
       insert(A[i],A,i)
    
def insert(v, A, hi):
    for i in range(hi-1,-1,-1):
        if v >= A[i]:
            A[i+1] = v
            return
        A[i+1] = A[i]
    A[0] = v
 
# recursive 
def mergeSort(A):
    if len(A) <= 1:
        return 
    mid = len(A)//2
    half1 = A[:mid]
    half2 = A[mid:]
    mergeSort(half1)
    mergeSort(half2)
    merge(half1,half2,A)        
 
def merge(h1, h2, A):
    i=0; j1=0; j2=0
    while j1<len(h1) and j2<len(h2):
        if h1[j1] < h2[j2]:
            A[i] = h1[j1]
            j1 += 1; i += 1
        else:
            A[i] = h2[j2]
            j2 += 1; i += 1
    while j1 < len(h1):
        A[i] = h1[j1]
        j1 += 1; i += 1
    while j2 < len(h2):
        A[i] = h2[j2]
        j2 += 1; i += 1
        
def quickSort(A):
    quickSortRec(A,0,len(A))
 
def quickSortRec(A, lo, hi):  
    if hi-lo <= 1:
        return
    iPivot = partition(A,lo,hi)
    quickSortRec(A,lo,iPivot)
    quickSortRec(A,iPivot+1,hi)
 
def partition(A, lo, hi):
    pivot = A[lo]
    B = [0 for i in range(lo,hi)]
    loB = 0
    hiB = len(B)-1
    for i in range(lo+1,hi):
        if A[i] < pivot:
            B[loB] = A[i]
            loB += 1
        else:
            B[hiB] = A[i]
            hiB -= 1
    B[loB] = pivot
    for i in range(len(B)):
        A[lo+i] = B[i]
    return lo+loB

# testing

def test():
    s_algs = (mergeSort, insertionSort)
    for i in s_algs:
        for j in range(1,6): # range amended to avoid timeout.
            print(sortTimeUsing(i,randomIntArray(10**j,10)))
        print("** next alg **")
test()
```

```python
# Question 7

def isSubArray(A,B):
    if len(B) > len(A):
        return False
    else:
        for v in range(len(B)):
            if equateArr(A,B,v):
                return True
        return False     
    
    
def equateArr(A,B,v):
    for i in range(len(B)):
        if A[i+v] != B[i]:
            return False
    return True

# testing

A = [5,6,7,8,9]
B = [6,7]
C = [5,8]

print(isSubArray(A,B)) 
print(isSubArray(A,C))
```
