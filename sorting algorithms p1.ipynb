```python
# Question 2
import time

# using generic selection sort functions from lecture.
def sortTime(A):
    t = time.time()
    for i in range(len(A)):
        imin = findMin(i,A)
        swap(i,imin,A)
    t = time.time()-t
    return t
        
def findMin(i, A):
    imin = i
    for j in range(i+1,len(A)):
        if A[j] < A[imin]:
            imin = j
    return imin
    
def swap(i, j, A):
    tmp = A[i]
    A[i] = A[j]
    A[j] = tmp
    
# testing

A = []
B = [4,8,12,2,3,56]
C = [1,2,3,4]

print("Time taken for A: ",sortTime(A))
print("A sorted is: ",A)

print("Time taken for B: ",sortTime(B))
print("A sorted is: ",B)

print("Time taken for C: ",sortTime(C))
print("C sorted is: ",C)
    
```

```python
# Question 4
import time

# sorting algorithms

def insertionSort(A):
    for i in range(1,len(A)):
        insert(A[i],A,i)
    
def insert(v, A, hi):
    for i in range(hi-1,-1,-1):
        if v >= A[i]:
            A[i+1] = v
            return
        A[i+1] = A[i]
    A[0] = v
    
def selectionSort(A):
    for i in range(len(A)):
        imin = findMin(i,A)
        swap(i,imin,A)
    
def findMin(i,A):
    imin = i
    for j in range(i+1, len(A)):
        if A[j] < A[imin]:
            imin = j
    return imin
    
def swap(i,j,A):
    tmp = A[i]
    A[i] = A[j]
    A[j] = tmp
    
## question function

def sortTimeUsing(sortf,A):
    t = time.time()
    sortf(A)
    t = time.time()-t
    return t

# testing

A = [56,3,1,3,4,5,66]
print("using insertion sort: ",sortTimeUsing(insertionSort,A))
print("using selection sort: ",sortTimeUsing(selectionSort,A))
```

```python
# Question 6

def insertionSortC(A):
    sortedA = []
    for i in range(len(A)):

        key = A[i]
        if len(sortedA)==0:
            sortedA.append(key)
        else:
            j = len(sortedA)-1
            while j > -2:
                if key >= sortedA[j]:
                    sortedA.insert(j+1,key)
                    j = -1
                j -= 1
                if j==-1:
                    sortedA.insert(0,key)
                    
    return sortedA
    
# testing

A = randomIntArray(10,10)
print("The original array: ",A)
print("The sorted array: ",insertionSortC(A))
```

```python
# Question 7

def insertionSort(A):
    sortedA = []
    for i in range(len(A)):
        insert(A[i],sortedA,i)
    return sortedA
    
def insert(v, A, hi):
    target_i = binSearch(A,v)
    A.insert(target_i+1,v)
     
def binSearch(A,v):    
    lo = 0
    hi = len(A)-1
    while (lo <= hi):
        mid = (lo+hi)//2
        if A[mid] >= v:
            return mid
        elif A[mid] < v:
            lo = mid+1
        else:
            hi = mid-1
    return -1

# testing

A = [1,1,1,3,4,5,7,10]
print(A)
print(insertionSort(A))

# Complexity is O(^2).
```
