```python
class ArrayList:
    def __init__(self):
        self.inArray = [0 for i in range(10)]
        self.count = 0
        
    def get(self, i):
        self._checkBounds(i, self.length())
        return self.inArray[i]

    def set(self, i, e):
        self._checkBounds(i, self.length())
        self.inArray[i] = e

    def length(self):
        return self.count

    def append(self, e):
        self.inArray[self.count] = e
        self.count += 1
        if len(self.inArray) == self.count:
            self._resizeUp()

    def insert(self, i, e):
        self._checkBounds(i, self.length())
        for j in range(self.count,i,-1):
            self.inArray[j] = self.inArray[j-1]
        self.inArray[i] = e
        self.count += 1
        if len(self.inArray) == self.count:
            self._resizeUp()
    
    def remove(self, i):
        self._checkBounds(i, self.length())
        self.count -= 1
        val = self.inArray[i]
        for j in range(i,self.count):
            self.inArray[j] = self.inArray[j+1]
        return val

    def _resizeUp(self):
        newArray = [0 for i in range(2*len(self.inArray))]
        for j in range(len(self.inArray)):
            newArray[j] = self.inArray[j]
        self.inArray = newArray
        
    def _checkBounds(self, i, hi):  # checks whether i is in [0,hi]
        if i < 0 or i > hi:
            raise Exception("index "+str(i)+" out of bounds!")
            
    def appendAll(self, A): # Question 1 
        for i in A:
            self.append(i)
                   
    def removeVal(self, e): # Question 2 
        for i in range(self.length()):
            if self.inArray[i] == e:
                self.remove(i)
                return True
        return False
    
    def toArray(self): # Question 3
        out = [0 for i in range(self.length())]
        for i in range(self.length()):
            out[i] = self.get(i)
        return out
            
    def palindrome(self): # Question 4
        newAr = [0 for i in range(len(self.inArray))]
        i = 0
        for j in range(self.length()-1,-1,-1):
            newAr[i] = self.inArray[j]
            i += 1
        self.inArray = newAr
        
    def sort(self): # Question 6
            
            for i in range(1,self.length()):
                targ = self.inArray[i]
                j = i - 1
                while j >= 0 and targ < self.inArray[j]:
                    self.inArray[j+1] = self.inArray[j]
                    j -= 1
                self.inArray[j] = targ
    
    def __str__(self):
        return str(self.inArray[:self.count])
```

```python
class PQElement:
    def __init__(self, v, p):
        self.val = v
        self.priority = p
        
    def __repr__(self):
        return "("+str(self.val)+","+str(self.priority)+")"
```

```python
class PQueue: 
    def __init__(self):
        self.inList = ArrayList()
  
    def size(self):
        return self.inList.length()
  
    def deq(self):
        return self.inList.remove(0)
    
    def enq(self,e):
        for i in range(self.inList.length()):
            if e.priority <= self.inList.get(i).priority:
                self.inList.insert(i,e)
                return
        self.inList.append(e)
        
    def __str__(self):
        return str(self.inList)
    
```
